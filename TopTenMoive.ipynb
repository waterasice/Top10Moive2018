```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
# 魔术方法表示在ipynb中显示图
%matplotlib inline
```

```python
# sort the values from the top to the least value and slice the first 5 items

film_data = {"Box office":[5.7,5.8,6.2,7.3,7.9,8.4,12.4,13,13.4,20.4],
            "Film name":[u'《红海行动》',u'《头号玩家》',u' 《蚁人2：黄蜂女现身》',u'《死侍2》',u'《碟中谍6》',u'《毒液》',u'《超人总动员2》',u'《侏罗纪世界2》',u'《黑豹》',u'《复仇者联盟3：无限战争》']}
newdf = pd.DataFrame(film_data)

```

```python
newdf
```

```python
newdf['Box office'].plot(kind = 'bar')
```

```python
fig,ax=plt.subplots()
ax.bar(newdf['Film name'],newdf['Box office'])
ax.set_xlabel("Year")  #设置x轴标签
ax.set_ylabel("Dogs Eaten")  #设置y轴标签
ax.set_title("2018年全球电影十大票房排行榜")  #设置标题
#ax.set_xlim(1979,2011)  #设置x轴数据限值
plt.xticks(rotation=45)
plt.show()  #显示图像
```
